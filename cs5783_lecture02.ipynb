```python
x = np.arange(11.)
c = (2*np.pi)/10.
print(c)
print(c*x)
x *= c
print(x)
y = np.sin(x)
```

```python
x = np.array([1,2,3,2,1])
y = np.array([1,3,2,3,1])
plt.subplot(2, 1, 1)
plt.plot(x)
plt.subplot(2, 1, 2)
plt.plot(y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)
# Add labels in plot command
plt.plot(np.sin(x), label='sin')
plt.plot(np.cos(x), label='cos')
plt.legend()
plt.show()
# Pass labels as a list to legend()
plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.legend(['sin','cos'])
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.title('It\'s a sine!', fontsize= 'large')
plt.show()
plt.plot(x, np.sin(x))
plt.grid()
plt.show()
```

```python
# Get the monochrome sample
# image from scipy
from scipy.datasets import ascent
img = ascent()

# Display image with bone
# colormap

plt.imshow(img, cmap=plt.cm.bone)

# Add a colorbar to the display
plt.colorbar()
plt.show()
```
